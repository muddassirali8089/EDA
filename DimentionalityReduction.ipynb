```python
import numpy as np
import pandas as pd
```

```python
ansur = pd.read_csv('ansur.csv', encoding='latin1')
ansur.head()

numeric_df = ansur.select_dtypes(include='number')
numeric_df.head()
```

```python
label = numeric_df['Weightlbs']  # Replace 'LabelName' with the actual column name
print(label.head())
numeric_df.head()
```

```python
numeric_df = numeric_df.drop('Weightlbs', axis=1)
numeric_df.head()


```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(numeric_df, label, test_size=0.2, random_state=42)

# Train the model
model = SVC()
model.fit(X_train, y_train)

# Predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print accuracies and the difference
print("Train Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
print("Difference (Train - Test):", train_accuracy - test_accuracy)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
label=ansur['Branch']
data = ansur.drop("Branch" , axis=1)
data = data.select_dtypes(include=['number'])


# Split data
X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=0.2, random_state=42)

# Train the model
model = SVC()
model.fit(X_train, y_train)

# Predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print accuracies and the difference
print("Train Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
print("Difference (Train - Test):", train_accuracy - test_accuracy)

```
